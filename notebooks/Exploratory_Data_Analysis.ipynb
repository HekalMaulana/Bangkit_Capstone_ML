## Import Library And Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/interests - Sheet1.csv")
df.head()
```

```python
df.info()
```

## Drop Kolom yang tidak digunakan dan mengurutkan kolom feature

```python
df = df.drop(columns=["6","16", "17"], axis=1)
df.info()
```

```python
kolom_baru = {
    '7': '6', '8': '7', '9': '8', '10': '9', '11': '10', 
    '12': '11', '13': '12', '14': '13', '15': '14'
}

df = df.rename(columns=kolom_baru)
df.info()
```

```python
feature_columns = [str(i) for i in range(1, 15)]
target_columns = ["y"]

column_order = feature_columns + target_columns

df = df[column_order]
```

```python
df.info()
```

## Menambahkan kolom baru 15-25 yang berisi nilai 0

```python
for i in range(15, 26):
    df[str(i)] = 0

df.info()
```

## Mengurutkan kembali kolom menjadi 1-25 dan y

```python
feature_columns = [str(i) for i in range(1, 26)]
target_columns = ["y"]

column_order = feature_columns + target_columns

df = df[column_order]
df.info()
```

## Check Missing Value

```python
print("Jumlah Missing Value :\n",df.isnull().sum())
```

## Simpan menjadi csv baru

```python
df.to_csv("../data/interests_without_onehot.csv", index=False, header=True)
```
