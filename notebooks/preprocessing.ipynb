# Preprocessing Data

## Import Library

```python
import pandas as pd
import numpy as np
```

## Import Dataset

```python
df = pd.read_csv("../data/interests - Sheet1.csv")
df.head()
```

## Drop Feature 16

```python
df = df.drop("16", axis=1)
df.head()
```

## Identifikasi Missing Value

```python
df.isnull().sum()
```

## Identifikasi Duplicate Value

```python
df.duplicated().sum()
```

### Melihat baris yang duplicate

```python
df[df.duplicated()]
```

### Menghapus baris yang duplicate

```python
df = df.drop_duplicates()
```

## Identifikasi ulang missing value dan duplicate value

```python
print("Jumlah Baris Duplicate : ", df.duplicated().sum())
print("Jumlah Baris Missing :\n", df.isnull().sum())
```

## Identifikasi jumlah unique value pada kolom target

```python
print("Jumlah unique value pada kolom target :", df["y"].nunique())
print(df["y"].unique())
```

### Transformasi kolom target (Kategori ke numerik) dengan

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
y = df["y"].values.reshape(-1, 1) # Kolom Target

onehot_encoder = OneHotEncoder(sparse_output=False)
y_onehot = onehot_encoder.fit_transform(y)

print("One-Hot Encoded :")
print(y_onehot)
```

### Tambahkan hasil onehot encode ke new dataframe

```python
# Buat dataframe untuk hasil onehot
onehot_columns = [f"y_{i}" for i in range(y_onehot.shape[1])]
df_onehot = pd.DataFrame(y_onehot, columns=onehot_columns)

# Drop kolom y pada df
df = df.drop(columns=["y"]).reset_index(drop=True)

# Gabungkan hasil encode ke df
new_df = pd.concat([df, df_onehot.reset_index(drop=True)], axis=1)

new_df.head()
```

## Identifikasi ulang missing value dan duplicate value

```python
print("Jumlah Baris Duplicate : ", new_df.duplicated().sum())
print("Jumlah Baris Missing :\n", new_df.isnull().sum())
```

# Informasi new_df

```python
new_df.info()
```

## Hapus kolom 17 dan urutkan kolom fitur dari 1-15

```python
new_df = new_df.drop(columns=["17"])

feature_colum = [str(i) for i in range(1, 16)]
target_column = [f"y_{i}" for i in range(7)]

column_order = feature_colum + target_column

new_df = new_df[column_order]
```

## Menyimpan new_df menjadi csv baru

```python
new_df.to_csv("../data/interest_new.csv", index=False)
```

## Analisis korelasi

```python
# Hitung korelasi antara fitur dan target
correlation_matrix = new_df.corr()

# Pilih hanya korelasi fitur (1-15) dengan target (y_0 hingga y_6)
feature_target_correlation = correlation_matrix.loc[[f"y_{i}" for i in range(7)], [str(i) for i in range(1, 16)]]

print(feature_target_correlation)

# Visualisasi dengan heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(feature_target_correlation, annot=True, cmap="coolwarm")
plt.title("Korelasi Fitur dengan Target")
plt.show()
```
